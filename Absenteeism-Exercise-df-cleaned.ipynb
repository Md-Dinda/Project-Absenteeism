## Converting Date Time to Weekday

### Transform the Date String to Date Time type

```python
########## 'Date' ################################
##################################################
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date']
```

```python
df_reason_mod.info()
```

```python
########## Extract the Month Value ###############
##################################################

list_months = []

for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
    
# list_months
# len(list_months)
df_reason_mod['Month Value'] = list_months
df_reason_mod.head(10)

```

```python
########## Month Value to Weekday ###############
##################################################
def date_to_weekday(date_value):
    return date_value.weekday()

df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head(5)
```

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
# df_reason_mod.head()
# df_reason_mod.columns.values
column_names_upd = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
df_reason_mod = df_reason_mod[column_names_upd]
df_reason_mod.head()
```

```python
# Membuat kolom untuk menghitung apakah karyawan absen (1 jika 'Absenteeism Time in Hours' > 0, 0 jika tidak)
raw_csv_data['Absenteeism'] = np.where(raw_csv_data['Absenteeism Time in Hours'] > 0, 1, 0)

# Mengelompokkan data berdasarkan 'Transportation Expense' dan 'Children'
grouped_df = raw_csv_data.groupby(['Transportation Expense', 'Children']).agg(
    total_absences=('Absenteeism', 'sum'),
    total_entries=('ID', 'count')
).reset_index()

# Menghitung probabilitas absen
grouped_df['Probability of Absenteeism'] = grouped_df['total_absences'] / grouped_df['total_entries']

# Menampilkan hasil
print(grouped_df)

# Plotting the results
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
sns.scatterplot(x='Transportation Expense', y='Children', size='Probability of Absenteeism', sizes=(20, 200), data=grouped_df, legend=False)
plt.title('Probability of Absenteeism by Transportation Expense & Number of Children')
plt.xlabel('Transportation Expense')
plt.ylabel('Number of Children')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assume 'df' is your dataframe with 'Transportation Expense' and 'Absenteeism Time in Hours' columns
df_new = df_cleaned.copy()
# Create a binary column for absenteeism (1 if absent, 0 if not)
# Assume 'df' is your dataframe with 'Transportation Expense' and 'Absenteeism Time in Hours' columns

# Create a binary column for absenteeism (1 if absent, 0 if not)
df_new['Absenteeism'] = np.where(df_new['Absenteeism Time in Hours'] > 0, 1, 0)

# Define bins for transportation expenses
bins = np.linspace(df_new['Transportation Expense'].min(), df_new['Transportation Expense'].max(), 10)
labels = [f'{int(bins[i])}-{int(bins[i+1])}' for i in range(len(bins)-1)]

# Bin the transportation expenses
df_new['Expense Bin'] = pd.cut(df_new['Transportation Expense'], bins=bins, labels=labels, include_lowest=True)

# Calculate the mean transportation expense for each bin
mean_expense = df_new.groupby('Expense Bin')['Transportation Expense'].mean()

# Calculate the probability of absenteeism for each bin
prob_absenteeism = df_new.groupby('Expense Bin')['Absenteeism'].mean()

# Plot the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(mean_expense, prob_absenteeism, color='#66b3ff')
plt.title('Probability of Absenteeism vs Transportation Expense')
plt.xlabel('Mean Transportation Expense')
plt.ylabel('Probability of Absenteeism')
plt.grid(True)
plt.show()
```
